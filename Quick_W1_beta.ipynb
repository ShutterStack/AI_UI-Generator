```python
import spacy
from spacy.training import Example
import random

# Load a blank English model
nlp = spacy.blank("en")

# Add the NER pipeline
ner = nlp.add_pipe("ner")

# Define labels
labels = ["DOMAIN", "ELEMENT", "USER_PERSONA"]
for label in labels:
    ner.add_label(label)

# Process training data and create training examples
train_data = []
for item in dashboard_data:
    text = item["text"]
    entities = item["entities"]
    annotations = {"entities": [(entity["start"], entity["end"], entity["label"]) for entity in entities]}
    train_data.append((text, annotations))

# Function to train the NER model
def train_ner_model(nlp, train_data, n_iter=30):
    optimizer = nlp.begin_training()
    for itn in range(n_iter):
        random.shuffle(train_data)
        losses = {}
        for text, annotations in train_data:
            doc = nlp.make_doc(text)
            example = Example.from_dict(doc, annotations)
            nlp.update([example], drop=0.35, losses=losses)
        print(f"Iteration {itn+1}: Losses - {losses}")

# Train the NER model
train_ner_model(nlp, train_data)

# Save the trained model
output_dir = "trained_ner_model"
nlp.to_disk(output_dir)
```
