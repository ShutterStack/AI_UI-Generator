```python
import json
import spacy
from spacy.training.example import Example
from spacy.scorer import Scorer
import random

# Load the data
with open('train_data.json', 'r') as f:
    train_data = json.load(f)
with open('val_data.json', 'r') as f:
    val_data = json.load(f)

# Initialize blank spaCy model
nlp = spacy.blank("en")

# Add NER component to pipeline
if "ner" not in nlp.pipe_names:
    ner = nlp.add_pipe("ner")
else:
    ner = nlp.get_pipe("ner")

# Add labels to NER component
for data in train_data:
    for ent in data["entities"]:
        ner.add_label(ent["label"])

# Prepare training data
train_examples = []
for data in train_data:
    entities = [(ent["start"], ent["end"], ent["label"]) for ent in data["entities"]]
    example = Example.from_dict(nlp.make_doc(data["text"]), {"entities": entities})
    train_examples.append(example)

# Prepare validation data
val_examples = []
for data in val_data:
    entities = [(ent["start"], ent["end"], ent["label"]) for ent in data["entities"]]
    example = Example.from_dict(nlp.make_doc(data["text"]), {"entities": entities})
    val_examples.append(example)

# Training loop with validation
optimizer = nlp.begin_training()
for epoch in range(30):
    losses = {}
    random.shuffle(train_examples)
    for example in train_examples:
        nlp.update([example], drop=0.5, losses=losses, sgd=optimizer)

    # Validation step
    val_losses = {}
    scorer = Scorer()
    for example in val_examples:
        pred_doc = nlp(example.text)
        example.predicted = pred_doc
        scorer.score([example])

    # Compute and print F-score
    scores = scorer.score(val_examples)
    f_score = scores.get('ents_f', None)
    print(f"Epoch {epoch} - Training Loss: {losses['ner']}, Validation F-score: {f_score}")

# Test the trained model
test_text = "I need a dashboard to monitor sales performance with a bar chart showing quarterly sales and a KPI widget for total sales and targets."
doc = nlp(test_text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
import json
import spacy
from spacy.training.example import Example
from spacy.scorer import Scorer
import random

# Load the data
with open('train_data.json', 'r') as f:
    train_data = json.load(f)
with open('val_data.json', 'r') as f:
    val_data = json.load(f)

# Initialize blank spaCy model
nlp = spacy.blank("en")

# Add NER component to pipeline
if "ner" not in nlp.pipe_names:
    ner = nlp.add_pipe("ner")
else:
    ner = nlp.get_pipe("ner")

# Add labels to NER component
for data in train_data:
    for ent in data["entities"]:
        ner.add_label(ent["label"])

# Prepare training data
train_examples = []
for data in train_data:
    entities = [(ent["start"], ent["end"], ent["label"]) for ent in data["entities"]]
    example = Example.from_dict(nlp.make_doc(data["text"]), {"entities": entities})
    train_examples.append(example)

# Prepare validation data
val_examples = []
for data in val_data:
    entities = [(ent["start"], ent["end"], ent["label"]) for ent in data["entities"]]
    example = Example.from_dict(nlp.make_doc(data["text"]), {"entities": entities})
    val_examples.append(example)

# Training loop with validation
optimizer = nlp.begin_training()
for epoch in range(30):
    losses = {}
    random.shuffle(train_examples)
    for example in train_examples:
        nlp.update([example], drop=0.5, losses=losses, sgd=optimizer)

    # Validation step
    val_losses = {}
    scorer = Scorer()
    for example in val_examples:
        pred_doc = nlp(example.text)
        example.predicted = pred_doc
        scorer.score([example])

    # Compute and print F-score (outside the loop)
    scores = scorer.score(val_examples)
    f_score = scores.get('ents_f', None)
    print(f"Epoch {epoch} - Training Loss: {losses['ner']}, Validation F-score: {f_score}")

# Test the trained model (optional)
test_text = "I need a dashboard to monitor sales performance with a bar chart showing quarterly sales and a KPI widget for total sales and targets."
doc = nlp(test_text)

for ent in doc.ents:
    print(ent.text, ent.label_)
```
